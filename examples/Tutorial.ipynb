```python
import sys
sys.path.append('..')

import numpy as np

from fa_kit import FactorAnalysis, BrokenStick

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def make_random_dataset(n_samp=10000, n_feat=100):
    """
    make some random data with correlated features
    """
    data = np.random.randn(n_samp, n_feat)
    
    signal_width = 10
    signal_overlap = 2
    step_size = signal_width - signal_overlap
    for i in range(0, data.shape[1], step_size):
        shared_signal = 0.4*np.random.randn(n_samp, 1)
        data[:, i:(i+signal_width)] += shared_signal
    return data

def plot_summary(fa, num_eigs_to_plot=30):
    """
    Plot a summary of the factor analysis
    """

    f = plt.figure(figsize=(8,8))
    ax_scree = plt.subplot(2, 1, 1)
    ax_raw_comp = plt.subplot(2, 2, 3)
    ax_rot_comp = plt.subplot(2, 2, 4)

    ax_scree.plot(fa.props_raw[:num_eigs_to_plot], '-ok')
    ax_scree.plot(fa.fit_stick.values[:num_eigs_to_plot], '--k')
    ax_scree.set_title('Eigenvalues with broken stick superimposed')

    ax_raw_comp.plot(fa.comps_raw[:, fa.retain_idx])
    ax_raw_comp.set_title('Raw Components')

    ax_rot_comp.plot(fa.comps_rot)
    ax_rot_comp.set_title('Rotated Components')

    return f
```

# Factor analysis on raw data

```python
data = make_random_dataset()
data_cov = data.T.dot(data)
```

```python
# Set up the factor analysis object, indiate whether the
# data is a square covariance matrix or a n_samples x n_feat data matrix
# with the bool `is_covar`

fa_raw = FactorAnalysis(
    data,
    is_covar=False
    )

# Extract the components
fa_raw.extract_components()

# Calculate how many components to retain
# You can use any of these methods:
#   'top_n', 'top_pct', 'kaiser', 'broken_stick'

fa_raw.find_comps_to_retain(method='broken_stick')

# Once you know how many to retain, re-extract with PAF
fa_raw.reextract_using_paf()

# Apply factor rotation
# Right now there are both 'varimax' and 'quartimax'
fa_raw.rotate_components(method='varimax')

# Plot summary
fig_raw = plot_summary(fa_raw, num_eigs_to_plot=30)
```
